```python
import pandas as pd
import plotly.express as px
```

```python
car_data = pd.read_csv('../vehicles_us.csv') # leer los datos
car_data.head()
```

```python
df_odometer = car_data.groupby(['odometer', 'condition'])['price'].mean().reset_index()
df_odometer
```

```python
fig = px.histogram(car_data, x="odometer", color='condition', nbins=50) # crear un histograma
fig.show() # crear grÃ¡fico de dispersiÃ³n
```

```python
fig = px.scatter(car_data, x="odometer", y="price") # crear un grÃ¡fico de dispersiÃ³n
fig.show() # crear grÃ¡fico de dispersiÃ³n
```

```python
fig = px.scatter(car_data, x="odometer", y="price", color='condition') # crear un grÃ¡fico de dispersiÃ³n odometer vs. price por condiciÃ³n.
fig.show() # crear grÃ¡fico de dispersiÃ³n
```

```python
df_model = car_data.groupby(['model_year', 'condition'])['price'].mean().reset_index()
df_model
```

```python
fig = px.line(df_model, x='model_year', y='price', color='condition', markers=True)
fig.show()
```
